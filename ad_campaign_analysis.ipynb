```python
import pandas as pd
import os
import pathlib
import matplotlib.pyplot as plt
from bokeh.plotting import ColumnDataSource, figure, show, output_file
from bokeh.io import output_notebook
from bokeh.models import HoverTool
from bokeh.models.formatters import DatetimeTickFormatter
plt.style.use('seaborn-deep')
%matplotlib inline
```

```python
cwd = os.chdir('gimbal')
```

```python
views = pd.read_excel('Data Analyst Assessment Data.xlsx', sheet_name=3)
views['str_date']=views['Date'].apply(lambda x: str(x)[:10])
checkouts = pd.read_excel('Data Analyst Assessment Data.xlsx', sheet_name=2)
checkouts['str_date']=checkouts['Date'].apply(lambda x: str(x)[:10])
```

```python
output_notebook() 
```

```python
tools_to_show = 'hover,box_zoom,pan,save,reset,wheel_zoom'
tips = [("Date", "@str_date"),("Facebook", "@Facebook"),("Organic", "@Organic"),("Youtube", "@Youtube")]
p = figure(title="Views", x_axis_label="Date", y_axis_label="Views (Mil)", x_axis_type="datetime", tools=tools_to_show, tooltips=tips)
p.line('Date','Facebook',source=views, legend="facebook", color='red')
p.line('Date','Organic',source=views, legend="organic", color='blue')
p.line('Date','Youtube',source=views, legend="youtube", color='gold')
p.legend.click_policy="hide"
p.legend.location="top_left"
hover = p.select(dict(type=HoverTool))
hover.tooltips = tips
hover.mode = 'mouse'
show(p)
```

```python
r = figure(title="Checkouts", x_axis_label="Date", y_axis_label="Checkouts (Mil)", x_axis_type="datetime", tools=tools_to_show, tooltips=tips)
r.line('Date','Facebook',source=checkouts, legend="facebook", color='red')
r.line('Date','Organic',source=checkouts, legend="organic", color='blue')
r.line('Date','Youtube',source=checkouts, legend="youtube", color='gold')
r.legend.click_policy="hide"
r.legend.location="top_left"
hover = r.select(dict(type=HoverTool))
tips = [("Date", "@str_date"),("Facebook", "@Facebook"),("Organic", "@Organic"),("Youtube", "@Youtube")]
hover.tooltips = tips
hover.mode = 'mouse'
tools_to_show = 'hover,box_zoom,pan,save,reset,wheel_zoom'
show(r)
```
